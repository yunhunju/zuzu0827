#### Data processing 
New final grade variable was created after data was loaded.
And then female and male data was stored in different variale. 

```python
df = pd.read_csv("student-mat.csv")

df['activities']=df['activities'].replace({'no': 0, 'yes': 1})
df['romantic']=df['romantic'].replace({'no': 0, 'yes': 1})
df['schoolsup']=df['schoolsup'].replace({'no': 0, 'yes': 1})
# replace no and yes to 0 and 1

```

```python
df['final'] = df['G1']+df['G2']+df['G3']
df['final'] = df['final']/3*5
df['final'] = df['final'].astype(int)
df
# Create new 'final variable.
```

```python
Fdf = df[df['sex'] == 'F']
Mdf = df[df['sex'] == 'M']

# Create seperated male and female dataset
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 120

plt.subplot(121)

label = ('No support', 'Support')
y_pos1 = np.arange(len(label))

Fdf.groupby(['schoolsup'])['final'].mean().plot.bar()
plt.ylim((0,60))
plt.xlabel('Female school support status')
plt.ylabel("Average grade")
plt.xticks(y_pos1, label)
plt.grid()

plt.subplot(122)

label2 = ('No support', 'Support')
y_pos2 = np.arange(len(label))

Mdf.groupby(['schoolsup'])['final'].mean().plot.bar()
plt.ylim((0,60))
plt.xlabel('Male school support status')
plt.ylabel("Average grade")
plt.xticks(y_pos2, label2)
plt.grid()
plt.tight_layout()

```

#### Context visualization 2 - School support vs Final grade
According to the result, students do not get support from school have higher fianl grade scores for both genders.
I can assume that students who do not have enough educational support from family or have difficulty keeping up with school course are eligible to receive additional school supports and their final grades were still affected by those lack of care.  

```python
newdf = df.groupby(['failures'],as_index = False).agg({'final': ['mean', 'count']})
sc_plot = plt.scatter(newdf['failures'], newdf['final']['mean'])
plt.xlabel('Failures', fontsize = 15)
plt.ylabel('Average grade', fontsize = 15)
plt.title('Number of failures & Average grade')
plt.xticks(np.arange(0,4,1))
sc_plot

# Scatter plot to see the relationship between failure and fianl grade. 
```

#### Context visualization 3 - Failure vs Final grade

As seen from the scatter plot graph, I can see previous failure experience(s) significantly affected final grades. I concluded that havitual behavior is not easily changed and therefore it can be considered predictory variable. 

```python
newdf = df.groupby(['studytime'],as_index = False).agg({'final': ['mean', 'count']})
sc_plot = plt.bar(newdf['studytime'], newdf['final']['mean'])
plt.xlabel('studytime/(hr)', fontsize = 15)
plt.ylabel('Average grade', fontsize = 15)
plt.title('Studytime & Average grade')
plt.xticks(np.arange(1,5,1))
plt.grid()
sc_plot

# Bar graph to see the relationship between studytime and average grade.

```

#### Context visualization 3 - Study time vs Final grade
Even if there is not so much noticeable grade increase between 1 to 2 and 3 to 4, I can see that study time still affect fianl grade. To be more specific, we can expect the average final grade from studying more than 10 hours to be higher than grade from less than 2 hours of study group.

```python
fig = plt.figure(figsize=(7,6))

plt.subplot(121)
bars = ('very low','low','moderate', 'high','very high')
y_pos = np.arange(len(bars))

Fdf.groupby(['goout'])['final'].mean().plot.bar()
plt.ylim((30,65))
plt.xlabel('female goout')
plt.ylabel("Average grade")
plt.xticks(y_pos, bars)
plt.grid()

plt.subplot(122)
bars = ('very low','low','moderate', 'high','very high')
y_pos = np.arange(len(bars))

Mdf.groupby(['goout'])['final'].mean().plot.bar()
plt.ylim((30,65))
plt.xlabel('male goout')
plt.ylabel("Average grade")
plt.xticks(y_pos, bars)
plt.grid()
plt.tight_layout()

# Bar graph to see the relationship between goout level and grade 
# Also, two bar graph was created to compare the results based on sex. 
```

#### Context visualization 4 - Go out vs Final grade
Going out level affected both genders but male group was more highly afftected by go out factor. 
Plus, some outlier values in female group with very low level of go out were found to afftect the result and therefore there is irregular trend. 

```python
df['pedu'] = df['Medu'] + df['Fedu']
df.describe()
# New variable pedu(parent education level) was created.
```

```python
newdf = df.groupby(['pedu'],as_index = False).agg({'final': ['mean', 'count']})
sc_plot = plt.bar(newdf['pedu'], newdf['final']['mean'], color = 'y')
plt.xlabel('parent education score', fontsize = 15)
plt.ylabel('Average grade', fontsize = 15)
plt.xticks(np.arange(1,9,1))
plt.title("Parent education & child's Average grade")

plt.grid()
sc_plot
```

### Building a multiple linear regresion modeling to find the most significant variable.

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(df[['failures','schoolsup','activities','romantic','goout','studytime']],df.final)
# Multiple linear regression to find the most significant variable and predict the final grade based on variables.
```

```python
reg.coef_
# coefficient values were earned by using multiple variables 
```

From the result, I fould failure and school support are the most significant variable in here. 

```python
reg.intercept_
# intercept value
```

```python
reg.predict([[3,1,0,1,4,1]])
```

As expected. the final score is quite low with those values.

```python
reg.predict([[0,0,1,0,1,4]])
```

As I expected with the values of no failures, without school support, doing activities, not in relationship, not so much going out, and long study hours, The predicted final score is higher than before.

## Interactivity work


```python
import bqplot
```

```python
Fx = Fdf['failures']
Fy = Fdf['final']

x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

x_ax = bqplot.Axis(scale=x_sc, label='Failure')
y_ax = bqplot.Axis(scale=y_sc, label='Grade', orientation='vertical')
```

```python
scatters = bqplot.Scatter(x=Fx, y=Fy, scales={'x':x_sc, 'y':y_sc})
```

```python
selector = bqplot.interacts.FastIntervalSelector( scale=x_sc, marks=[scatters])

scatters.unselected_style = {'opacity': 0.4}
scatters.selected_style = {'fill':'yellow', 'stroke':'blue'}

fig = bqplot.Figure(marks=[scatters], axes=[x_ax,y_ax], interaction = selector)
display(fig)

# I tried using interactivity so that the result can highligt the selected ares.
```

From the result, I can check students who did not fail before are the majority their final grade is widely spreaded.
At the same time, student who did fail 3 times already are only 6 from total students. Also, I can see more failure numbers lead to lower grade result.
